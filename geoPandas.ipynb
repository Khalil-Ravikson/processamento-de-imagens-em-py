```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem (exemplo com uma imagem de satlite)
img = cv2.imread(r'D:\EDA\py\imagePy\onde-fazer-o-download-de-modelos-digitais-de-elevacao.jpg')

# Aplicar o filtro bilateral
img_bilateral = cv2.bilateralFilter(img, d=15, sigmaColor=75, sigmaSpace=75)

# Exibir a imagem original e a suavizada
plt.figure(figsize=(10,5))

# Imagem original
plt.subplot(1, 2, 1)
plt.title('Imagem Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Imagem com filtro bilateral
plt.subplot(1, 2, 2)
plt.title('Imagem com Filtro Bilateral')
plt.imshow(cv2.cvtColor(img_bilateral, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
